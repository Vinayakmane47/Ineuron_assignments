#### Q1.Write a Python Program to Check if a Number is Positive, Negative or Zero?

```python
def check(n): 
    if type(n)==int : 
        if n <0 : 
            print("Negative Number")
        elif n > 0 : 
            print("Positive Number")
        else : 
            print("Zero")
            
    else : 
        print("Please enter a integer value")
check(10)
```

#### Q2.Write a Python Program to Check if a Number is Odd or Even?

```python
def check1(n):  
    try : 
        if type(n) == int:
            if n%2 == 0 : 
                print("{} is an even number".format(n))
            else : 
                print("{} is an odd number".format(n))
        else : 
            print("Please Enter an integer ")
                
    except Exception as e : 
        print(e)
        
check1(10)
            
```

#### Q3.Write a Python Program to Check Leap Year?

```python
def leap(n):
    if type(n)==int : 
        if n%4 == 0 : 
            if n%100 == 0: 
                if n%400 == 0: 
                    print("The year {} is a leap year".format(n))
                else : 
                    print("The year {} is a not a  leap year, it has 365 days ".format(n))
                    
            else : 
                print("The year {} is a leap year".format(n))
        else : 
            print("The year {} is a not a  leap year, it has 365 days ".format(n))
            
leap(1999)    
```

#### Q4.Write a Python Program to Check Prime Number?

```python
def prime(n): 
    if type(n) == int : 
        for i in range(2,n): 
            if n%i == 0: 
                print("{} is not a Prime Number".format(n))
                break 
        else : 
            print("{} is a prime Number".format(n))
            
prime(11)            
```

#### Q5.Write a Python Program to Print all Prime Numbers in an Interval of 1-10000?

```python
def prime_range(n): 
    if type(n)==int: 
        for i in range(2,n): 
            for j in range(2,i): 
                if i%j == 0: 
                    break 
            else : 
                print("{} is a Prime Number ".format(i))
prime_range(10000)                
```
